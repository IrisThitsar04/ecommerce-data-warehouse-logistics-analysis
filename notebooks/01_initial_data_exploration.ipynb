# 01_Initial Data Exploration

This notebook performs the initial exploration of the raw Olist e-commerce datasets to understand their structure, quality, and key characteristics prior to transformation.

The analysis includes:

- **Loading and inspecting raw datasets**  

- **Dataset structure analysis**, reviewing row counts, column definitions, and data types  

- **Missing data and duplicate detection**, identifying potential data quality issues  

- **Value distribution analysis**, examining ranges and unique values for key variables  

- **Preliminary data quality assessment**, identifying anomalies and inconsistencies that may impact downstream processing  

The goal of this step is to establish a clear understanding of the raw data and highlight data quality considerations before proceeding to data preparation, validation, and warehouse loading.

**Notes:**

**order_approved_at** : 160 missing

This could occur when:

- Payment was not approved

- The order was cancelled

- The order was expired or failed

Therefore, missing approval timestamps likely correspond to orders that did not progress beyond the payment stage, rather than indicating a data quality issue.

**order_delivered_carrier_date** : 1,783 missing

This could occur when:

- The order was cancelled before shipment

- The order was never handed to a carrier

- The logistics process did not reach the dispatch stage

These missing values likely reflect orders that did not progress to the shipping stage.

**order_delivered_customer_date** : 2,965 missing

This could occur when:

- Orders were cancelled

- Orders were still in transit at the time of data collection

- Orders were undeliverable

In such cases, the delivery process may not have been completed, which explains the missing timestamps.

```python
#count orders where approval timestamp is missing but the carrier delivery date exists
(df_orders['order_approved_at'].isna()& df_orders['order_delivered_carrier_date'].notna()).sum()
```

**Notes:**

Out of **99,441 orders**, **14 records** contain a delivery carrier timestamp despite a missing approval timestamp.

Given the very small proportion, this likely represents rare edge cases or logging inconsistencies rather than a systemic data quality issue.


```python
#count orders where approval timestamp is missing but the customer delivery date exists
(df_orders['order_approved_at'].isna()& df_orders['order_delivered_customer_date'].notna()).sum()
```

**Notes:**

Across the orders dataset (**99,441 records**), **14 orders** were identified where delivery timestamps
exist despite a missing approval timestamp.

These same records were consistently observed at both
carrier handoff and customer delivery stages, indicating isolated logging gaps rather than systemic
data quality issues.

```python
#count orders where the carrier delivery date is missing but the customer delivery date exists
(df_orders['order_delivered_carrier_date'].isna()& df_orders['order_delivered_customer_date'].notna()).sum()
```

**Notes:**

A single order exhibits a customer delivery timestamp without a recorded carrier handoff timestamp.

As it is 1 out of 99,441 orders, this is treated as a rare logging anomaly.

Validation showed that some review IDs appear more than once, indicating that the raw dataset allows duplicate review records.

The uniqueness constraint was therefore removed, and `review_id` is validated for non-null values only.  
This behaviour is treated as a characteristic of the source data rather than a data quality error.

Uniqueness may be enforced after data cleaning or deduplication in future work.

```python
def build_dim_date(df_orders, df_order_item):
  date_cols_orders=[
      'order_purchase_timestamp',
      'order_approved_at',
      'order_delivered_carrier_date',
      'order_delivered_customer_date',
      'order_estimated_delivery_date',
  ]

  date_cols_items=['shipping_limit_date']

  dates=[]

  for c in date_cols_orders:
    if c in df_orders.columns:
      dates.append(df_orders[c].dt.date)

  for c in date_cols_items:
    if c in df_order_item.columns:
      dates.append(df_order_item[c].dt.date)

  all_dates=pd.concat(dates).dropna().unique()
  df=pd.DataFrame({'full_date':pd.to_datetime(all_dates)})
  df = df.drop_duplicates(subset=['full_date'])
    
  df['date_sk']=df['full_date'].dt.strftime('%Y%m%d').astype(int)
  df['day']=df['full_date'].dt.day
  df['month']=df['full_date'].dt.month
  df['month_name']=df['full_date'].dt.strftime('%B')
  df['quarter']=df['full_date'].dt.quarter
  df['year']=df['full_date'].dt.year
  df['day_of_week']=df['full_date'].dt.weekday+1
  df['is_weekend']=df['day_of_week'].isin([6, 7])

  return df.sort_values('full_date')
```

```python
def build_dim_geography(df_geolocation):

  def mode_or_null(series):
    s=series.dropna()
    if s.empty:
      return None
    return s.value_counts().idxmax()

  dim_geo_df=(
      df_geolocation
      .groupby('geolocation_zip_code_prefix', as_index=False)
      .agg(
          city=('geolocation_city', mode_or_null),
          state=('geolocation_state', mode_or_null),
          lat=('geolocation_lat', 'mean'),
          lng=('geolocation_lng', 'mean')
      )
      .rename(columns={'geolocation_zip_code_prefix':'zip_code_prefix'})
  )
  return dim_geo_df
```

```python
def build_dim_customer(customers_df, dim_geo_df):

    dim_customer_df=(
        customers_df.loc[:, [
            'customer_unique_id',
            'customer_zip_code_prefix',
            'customer_city',
            'customer_state'
        ]]
        .drop_duplicates(subset=['customer_unique_id'])
        .merge(
            dim_geo_df[['geo_sk','zip_code_prefix']],
            left_on='customer_zip_code_prefix',
            right_on='zip_code_prefix',
            how='left'
        )
        .rename(columns={'geo_sk':'customer_geo_sk'})
        .drop(columns=['zip_code_prefix'])
        .reset_index(drop=True)
    )
    return dim_customer_df
```

```python
def build_fact_orders(df_orders, df_customers, engine):

    # Load dimension tables
    customer_dim = pd.read_sql(
        "SELECT customer_sk, customer_unique_id FROM dim.dim_customer",
        engine
    )

    date_dim = pd.read_sql(
        "SELECT date_sk, full_date FROM dim.dim_date",
        engine
    )

    date_dim["full_date"] = pd.to_datetime(date_dim["full_date"]).dt.date

    customer_lookup = df_customers[
        ["customer_id", "customer_unique_id"]
    ]

    fact_df = df_orders.merge(
        customer_lookup,
        on="customer_id",
        how="left"
    )

    fact_df = fact_df.merge(
        customer_dim,
        on="customer_unique_id",
        how="left"
    )

    date_cols = [
        "order_purchase_timestamp",
        "order_approved_at",
        "order_delivered_carrier_date",
        "order_delivered_customer_date",
        "order_estimated_delivery_date"
    ]

    for col in date_cols:
        fact_df[col] = pd.to_datetime(fact_df[col]).dt.date

    def resolve_date_sk(df, date_col, sk_name):
        df = df.merge(
            date_dim,
            left_on=date_col,
            right_on="full_date",
            how="left"
        )

        df = df.rename(
            columns={"date_sk": sk_name}
        ).drop(columns=["full_date"])

        return df

    fact_df = resolve_date_sk(
        fact_df,
        "order_purchase_timestamp",
        "purchase_date_sk"
    )

    fact_df = resolve_date_sk(
        fact_df,
        "order_approved_at",
        "approved_date_sk"
    )

    fact_df = resolve_date_sk(
        fact_df,
        "order_delivered_carrier_date",
        "carrier_handoff_date_sk"
    )

    fact_df = resolve_date_sk(
        fact_df,
        "order_delivered_customer_date",
        "delivered_customer_date_sk"
    )

    fact_df = resolve_date_sk(
        fact_df,
        "order_estimated_delivery_date",
        "estimated_delivery_date_sk"
    )

    fact_df["delivery_duration_days"] = (
        pd.to_datetime(fact_df["order_delivered_customer_date"]) -
        pd.to_datetime(fact_df["order_purchase_timestamp"])
    ).dt.days

    delay = (
        pd.to_datetime(fact_df["order_delivered_customer_date"]) -
        pd.to_datetime(fact_df["order_estimated_delivery_date"])
    ).dt.days

    fact_df["delivery_delay_days"] = delay.apply(
        lambda x: x if pd.notnull(x) and x > 0 else 0
    )

    fact_df["is_delivered_flag"] = fact_df["order_status"].eq("delivered")

    fact_df["is_late_delivery"] = fact_df["delivery_delay_days"] > 0

    fact_df["is_failed_order"] = ~fact_df["order_status"].eq("delivered")

    fact_df["issue_flag"] = (
        fact_df["is_late_delivery"] |
        fact_df["is_failed_order"]
    )

    final_cols = [
        "order_id",
        "customer_id",
        "customer_sk",
        "purchase_date_sk",
        "approved_date_sk",
        "carrier_handoff_date_sk",
        "delivered_customer_date_sk",
        "estimated_delivery_date_sk",
        "order_status",
        "delivery_duration_days",
        "delivery_delay_days",
        "is_delivered_flag",
        "is_late_delivery",
        "is_failed_order",
        "issue_flag"
    ]

    return fact_df[final_cols]
```

```python
def build_fact_order_items(df_order_items, engine):

    product_dim = pd.read_sql(
        "SELECT product_id, product_sk FROM dim.dim_product",
        engine
    )

    seller_dim = pd.read_sql(
        "SELECT seller_id, seller_sk FROM dim.dim_seller",
        engine
    )

    date_dim = pd.read_sql(
        "SELECT date_sk, full_date FROM dim.dim_date",
        engine
    )

    fact_orders_lookup = pd.read_sql(
        """
        SELECT order_id, customer_sk, purchase_date_sk
        FROM fact.fact_orders
        """,
        engine
    )

    date_dim["full_date"] = pd.to_datetime(date_dim["full_date"]).dt.date

    # Start with order_items
    fact_df = df_order_items.copy()

    # Product SK
    fact_df = fact_df.merge(
        product_dim,
        on="product_id",
        how="left"
    )

    # Seller SK
    fact_df = fact_df.merge(
        seller_dim,
        on="seller_id",
        how="left"
    )

    # Customer SK + purchase_date_sk
    fact_df = fact_df.merge(
        fact_orders_lookup,
        on="order_id",
        how="left"
    )

    # Shipping limit date → date_sk
    fact_df["shipping_limit_date"] = pd.to_datetime(
        fact_df["shipping_limit_date"]
    ).dt.date

    fact_df = fact_df.merge(
        date_dim,
        left_on="shipping_limit_date",
        right_on="full_date",
        how="left"
    ).rename(columns={"date_sk": "shipping_limit_date_sk"}).drop(columns=["full_date"])

    # Measure
    fact_df["item_total_value"] = (
        fact_df["price"] + fact_df["freight_value"]
    )

    final_cols = [
        "order_id",
        "order_item_id",
        "product_sk",
        "seller_sk",
        "customer_sk",
        "purchase_date_sk",
        "shipping_limit_date_sk",
        "price",
        "freight_value",
        "item_total_value"
    ]

    return fact_df[final_cols]
```

```python
def build_fact_seller_order(df_order_items, engine):
    seller_dim=pd.read_sql(
        'SELECT seller_sk, seller_id FROM dim.dim_seller',
        engine
    )
    fact_orders=pd.read_sql(
        '''
        SELECT
            order_id,
            customer_sk,
            purchase_date_sk,
            is_delivered_flag,
            is_late_delivery,
            is_failed_order,
            delivery_delay_days
        FROM fact.fact_orders
        ''',
        engine
    )

    agg=(
        df_order_items
        .groupby(['order_id', 'seller_id'], as_index=False)
        .agg(
            seller_order_revenue=('price', 'sum'),
            seller_order_freight=('freight_value', 'sum'),
            seller_order_items_count=('order_item_id', 'count')
        )
    )
    agg=agg.merge(seller_dim, on='seller_id', how='left')

    fact_df=agg.merge(fact_orders, on='order_id', how='left')

    fact_df["is_late_delivery"] = fact_df["is_late_delivery"].fillna(False)
    fact_df["is_failed_order"] = fact_df["is_failed_order"].fillna(False)
    fact_df["is_delivered_flag"] = fact_df["is_delivered_flag"].fillna(False)
    fact_df["delivery_delay_days"] = fact_df["delivery_delay_days"].fillna(0)

    final_cols=[
        'order_id',
        'seller_sk',
        'customer_sk',
        'purchase_date_sk',
        'seller_order_revenue',
        'seller_order_freight',
        'seller_order_items_count',
        'is_late_delivery',
        'is_delivered_flag',
        'is_failed_order',
        'delivery_delay_days'
    ]
    return fact_df[final_cols]
```
